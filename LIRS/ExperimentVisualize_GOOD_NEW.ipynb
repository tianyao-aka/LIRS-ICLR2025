```python
import numpy as np
import matplotlib.pyplot as plt

def plot_data_with_std_markers(data_str):
    
    plt.rcParams['font.size'] = 30
    plt.rcParams['font.family'] = 'serif'
    """
    Plots multiple lines with shaded standard deviation areas and different markers for each line.
    
    Parameters:
    - data_str (str): Multi-line string where each line represents a set of measurements
                      with means and standard deviations (meanstd).
    
    The function parses the string, converts it into numerical data, and plots each column as a separate line
    with distinct markers. The shaded region around each line represents the standard deviation.
    """
    # Split the data into lines and then into individual entries
    data = data_str.strip().split("\n")
    means = []
    std_devs = []
    
    # Parse each line
    for line in data:
        line_means = []
        line_stds = []
        for entry in line.split():
            mean, std = entry.split("")
            line_means.append(float(mean))
            line_stds.append(float(std))
        means.append(line_means)
        std_devs.append(line_stds)
    
    means = np.array(means)
    std_devs = np.array(std_devs)
    
    # Custom x-axis values
    # x = np.array([0.1, 0.3, 0.5, 0.75, 0.85, 0.9])
    x = np.array([30,50,70,90])
    
    # Different markers for each line
    markers = ['o', '^', 's', 'p']  # Example: circle, triangle_up, square, pentagon
    
    # Plotting
    plt.figure(figsize=(15, 12))
    
    # Plot each column
    name = ['SPMotif-0.40','SPMotif-0.60','SPMotif-0.70']
    for i in range(means.shape[1]):
        plt.plot(x, means[:, i], marker=markers[i], label=name[i])
        plt.fill_between(x, means[:, i] - std_devs[:, i], means[:, i] + std_devs[:, i], alpha=0.15)
    
    plt.xlabel(r'Lowest $K\%$ probability edges',fontsize=24)
    # plt.xlabel(r'Edge budget $\eta$',fontsize=24)
    plt.ylabel('Test metrics',fontsize=24)
    plt.legend(fontsize=24,loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    # plt.show()
    plt.savefig('ttt_experiments/Experiment_Figures/ablation_edge_div_spmotif_v2.pdf')
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import ast

def get_best_mean_for_combinations(file_path, columns, target_col="mean"):
    """
    Load an Excel file and return the best (maximum) value in the target column ('mean') for each unique combination
    of values in the specified columns. Handles merged cells by filling NaN values with the preceding value in the column.

    Args:
        file_path (str): The path to the Excel file.
        columns (tuple): A tuple of two column names to use for generating unique combinations.
        target_col (str): The column name to evaluate for the best value (default is 'mean').

    Returns:
        dict: A dictionary where keys are 2-tuples of the unique combinations and values are the best values in the target column.
    """
    # Load the Excel file into a DataFrame
    df = pd.read_csv(file_path)
    # Fill NaN values by forward filling with the preceding value in each column
    df.fillna(method='ffill', inplace=True)
    # df = df[df['edgePenalty']>=1]
    if 'list' in df.columns:
        df['list'] = df['list'].apply(ast.literal_eval)
        df = df[df['list'].apply(len) >= 3]
    if 'epoch' in df.columns:
        df['epoch'] = df['epoch'].apply(int)
        df = df[df['epoch'] > 10]
    # Initialize an empty dictionary to store the results
    result_dict = {}
    # print (df.head(5))
    # Get the unique combinations of values from the specified columns
    unique_combinations = df.drop_duplicates(subset=columns)

    # Iterate through each unique combination
    for _, combination in unique_combinations.iterrows():
        # Filter the DataFrame for rows matching the current combination
        filtered_df = df[(df[columns[0]] == combination[columns[0]]) & (df[columns[1]] == combination[columns[1]])]
        # Find the maximum value in the target column
        best_value = filtered_df[target_col].max()
        
        # Store the combination as a tuple and the best value in the dictionary
        result_dict[(combination[columns[0]], combination[columns[1]])] = best_value
    return result_dict


def generate_heatmap_from_dict(data_dict, x_label, y_label, fontsize=32,save_name = None):
    """
    Generates a heatmap from a dictionary, with customizable axis labels and font size.

    Args:
        data_dict (dict): A dictionary where keys are 2-tuples representing coordinates, and values are the heatmap values.
        x_label (str): The label for the x-axis.
        y_label (str): The label for the y-axis.
        fontsize (int): The font size for the heatmap annotations and labels. Default is 32.
    """
    # Convert dictionary to a DataFrame
    df = pd.DataFrame.from_dict(data_dict, orient='index', columns=['Value'])
    df.index = pd.MultiIndex.from_tuples(df.index, names=[y_label, x_label])
    df = df.unstack()  # Unstack to create a 2D matrix for heatmap

    # Flatten the column multi-index created by unstacking
    df.columns = df.columns.droplevel(0)
    
    # Format the values to xx.xx
    df = df.applymap(lambda x: round(x * 100, 2))

    # Plot the heatmap
    plt.figure(figsize=(10, 8))
    sns.set(font_scale=fontsize / 12.8)  # Adjust font scale to match the fontsize argument
    ax = sns.heatmap(df, annot=True, fmt=".2f", cmap="YlGnBu", cbar=False, linewidths=.5, annot_kws={"size": fontsize})

    # Adjust the axes labels
    ax.set_xlabel(x_label, fontsize=fontsize)
    ax.set_ylabel(y_label, fontsize=fontsize)

    # Show the heatmap
    plt.tight_layout()
    if save_name is None:
        plt.show()
    else:
        plt.savefig(save_name)


# generate_heatmap_from_dict(data_dict, x_label="edgePenalty", y_label="edgeProbThres", fontsize=32)

```

# Merge multiple PDF into one for motif visualze

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import os
from matplotlib.image import imread
from math import ceil

def merge_pngs_from_directory(input_dir, output_pdf, figs_per_row=2):
    """
    Merges all PNG files in a specified directory into a single PDF while maintaining the aspect ratio 
    and automatically adjusting the layout.

    Args:
        input_dir (str): Path to the directory containing PNG files to be merged.
        output_pdf (str): Path to the output merged PDF file.
        figs_per_row (int): Number of figures per row in the merged PDF. Defaults to 2.
    """
    # Get a sorted list of all PNG files in the directory
    png_list = sorted([os.path.join(input_dir, f) for f in os.listdir(input_dir) if f.endswith('.png')])
    
    num_pngs = len(png_list)
    num_rows = ceil(num_pngs / figs_per_row)  # Calculate the number of rows needed
    
    # Create a new PDF with multiple subplots
    with PdfPages(output_pdf) as pdf:
        for i in range(0, num_pngs, figs_per_row):
            fig, axes = plt.subplots(nrows=1, ncols=figs_per_row, figsize=(11, 8.5))
            axes = axes.flatten()
            
            for j in range(figs_per_row):
                idx = i + j
                if idx < num_pngs:
                    # Read the PNG image
                    img = imread(png_list[idx])
                    
                    # Calculate the aspect ratio of the image
                    aspect_ratio = img.shape[1] / img.shape[0]
                    
                    # Adjust the size of the subplot to maintain the aspect ratio
                    axes[j].set_aspect(aspect_ratio)
                    
                    # Display the image
                    axes[j].imshow(img)
                    axes[j].axis('off')  # Turn off the axis
                else:
                    axes[j].remove()  # Remove empty subplots
            
            # Adjust layout to prevent overlap
            plt.tight_layout()
            
            # Save the current figure into the PDF
            pdf.savefig(fig)
            plt.close(fig)



# Example usage:
input_directory = "plot_viz/basis/"
output_file = "plot_viz/basis.pdf"

# Merge the PDFs from the directory
merge_pngs_from_directory(input_directory, output_file)
```
